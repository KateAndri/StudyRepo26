1.   Загружаю библиотеки



```python
import pandas as pd
!pip install -U autogluon
from autogluon.tabular import TabularDataset, TabularPredictor
```

2.   Подгружаю данные



```python
data = TabularDataset(r'/content/dispensarization_data_2026.csv')
```

3. Оцениваю размер данных

```python
print(f"Размер данных: {data.shape}")
```

4. Смотрю первые 5 строк для оценки состояния

```python
print(f"\nПервые 5 строк:")
print(data.head())
```

5. Распределение целевой переменной

```python
print("Распределение 'Доклинический_риск':")
print(data['Доклинический_риск'].value_counts())
print(f"\nДоля единиц: {data['Доклинический_риск'].mean():.2%}")
```

6. Делю на выборки

```python
# Обучающая выборка — 80%, тестовая — 20%
train_data = data.sample(frac=0.8, random_state=42)
test_data = data.drop(train_data.index)

print(f"Обучающая выборка: {len(train_data)} строк")
print(f"Тестовая выборка: {len(test_data)} строк")
print(f"\nРаспределение в обучении:")
print(train_data['Доклинический_риск'].value_counts(normalize=True))
```

7. Создаем predictor и запускаю обучение моделей

```python
predictor = TabularPredictor(
    label='Доклинический_риск',
    eval_metric='roc_auc',
    verbosity=2
)

print("Начинаем обучение...")
predictor.fit(train_data)
print("Обучение завершено!")
```

8. Оценка на тестовой выборке

```python
performance = predictor.evaluate(test_data)

print("\nРезультаты на тестовой выборке:")
for metric, score in performance.items():
    print(f"  {metric}: {score:.4f}")
```

```python
leaderboard = predictor.leaderboard(test_data, silent=True)
print("\nСравнение всех моделей:")
print(leaderboard[['model', 'score_test', 'pred_time_test', 'fit_time']].head(10))
```

Модель показала отличные результаты на тестовой выборке (200 пациентов):

ROC-AUC = 0.98 — модель отлично различает пациентов с риском и без него

Accuracy = 98% — правильно предсказано 196 из 200 случаев

F1-score = 0.89 — хороший баланс между точностью и полнотой

Recall = 0.85 — модель выявляет 85% реальных случаев риска

Модель успешно справляется с сильным дисбалансом классов (всего 6% положительных примеров) и готова к практическому применению.

9. Определяем важность признаков

```python
feature_importance = predictor.feature_importance(test_data)

feature_importance_sorted = feature_importance.sort_values('importance', ascending=False)

print("\n Важность признаков для модели:")
print(feature_importance_sorted[['importance', 'stddev']].round(4))
```

Выделены три ключевых фактора доклинического риска:

Глюкоза_натощак_ммоль_л — самый важный признак (важность 0.105)

ИМТ (индекс массы тела) — второй по значимости (важность 0.055)

ЛПНП_ммоль_л (плохой холестерин) — третий значимый фактор (важность 0.024)

Остальные признаки (курение, пол, артериальное давление и др.) показали близкую к нулю важность, что может указывать на их меньшую прогностическую ценность в данной модели.
